```python
df = pd.read_csv("olist_products_dataset.csv")
```

```python
categorias = pd.DataFrame(df['product_category_name'].tolist(),columns=['Categorias'])
```

```python
print(categorias)
```

```python
categorias['Categorias'].value_counts().sort_values(ascending=False).head(10).plot(kind='barh', title='As categorias do produtos mais vendido', figsize=(10,15), fontsize=14, legend=True, color='green')
```

```python
df1 = pd.read_csv("olist_sellers_dataset.csv")
```

```python
cidades = pd.DataFrame(df1['seller_city'].tolist(),columns=['Cidades'])
```

```python
cidades['Cidades'].value_counts().sort_values(ascending=False).head(20).plot.pie(figsize=(10, 910))#.plot(kind='barh', title='As cidades que mais vendem', figsize=(10,15), fontsize=14, legend=True, color='orange')
```
